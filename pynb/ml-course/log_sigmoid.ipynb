```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def sigmoid(x):
    s = 1 / ( 1 +  np.exp(-x))
    return s

def sigmoid_derivative(x):
    s = sigmoid(x)
    ds = s * ( 1- s)
    return ds
```

```python
values = np.linspace(-10, 10, 100)
plt.plot(values, sigmoid(values), 'r')
plt.plot(values, sigmoid_derivative(values), 'b')

plt.grid()
plt.show()
```
